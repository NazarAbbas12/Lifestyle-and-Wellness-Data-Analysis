```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv('ScreenTime vs MentalWellness.csv')
```

```python
sns.scatterplot(x='stress_level_0_10',y='productivity_0_100',data=df ,hue="stress_level_0_10",
                 palette="coolwarm",alpha=0.7, s=70)
plt.xlabel('Stress Level(0-10)')
plt.ylabel('Productivity(0-100)')
plt.title('Relation between Stress and Productivity',fontsize=14)
plt.legend().set_visible(False)
plt.show()
```

```python
sns.scatterplot(data=df,x='stress_level_0_10',y='exercise_minutes_per_week',hue='stress_level_0_10',
                palette='inferno',alpha=0.7,s=70)
plt.xlabel('Stress Level (0-10)')
plt.ylabel('Exercise per week (Minutes)')
plt.title("Stress vs Exercise Relationship",fontsize=14)
plt.legend().set_visible(False)
plt.show()
```

```python
sns.scatterplot(data=df,x='stress_level_0_10',y='mental_wellness_index_0_100',hue='stress_level_0_10',
                palette='viridis',alpha=0.7,s=70)
plt.xlabel('Stress Level (0-10)')
plt.ylabel('Mental Wellness Index (0-100)')
plt.title("Stress vs Mental Wellness Index",fontsize=14)
plt.legend().set_visible(False)
plt.savefig("relation_stress_mental_wellness.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
sns.lineplot(data=df,x='age',y='productivity_0_100',color='royalblue',linewidth=2.5,
             marker='o',markersize=6)
plt.xlabel('Age')
plt.ylabel('Productivity (0-100)')
plt.title('Age vs Productivity', fontsize=14)
plt.grid(True,linestyle='--',alpha=0.6)
plt.savefig("age_VS_prod.png", dpi=300, bbox_inches='tight')
plt.show()
```
