```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv('ScreenTime vs MentalWellness.csv')
```

```python
df.info()
```

```python
plt.hist(df['age'], 
         bins=10, 
         color='skyblue', 
         edgecolor='black', 
         alpha=0.8)
plt.xlabel('Age Group')
plt.ylabel('Number of Persons')
plt.title('Age Distribution of Participants',fontsize=14)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
df_mean_sleep = df.groupby('gender')['sleep_hours'].mean().reset_index()
sns.barplot(data=df_mean_sleep,x='gender',y='sleep_hours',hue='sleep_hours',palette='Greens')
plt.xlabel('Gender')
plt.ylabel('Average Sleep Time(hours)')
plt.legend().set_visible(False)
plt.title('Average Sleep Time of genders',fontsize=14)
plt.ylim(0,8)
plt.show()
```

```python
df_mean_screen = df.groupby('gender')['screen_time_hours'].mean().reset_index()
sns.barplot(data=df_mean_screen,x='gender',y='screen_time_hours',hue='screen_time_hours',palette='Purples')
plt.xlabel('Gender')
plt.ylabel('Average Screen Time(hours)')
plt.legend().set_visible(False)
plt.title('Average Screen Time of genders',fontsize=14)
plt.ylim(0,10)
plt.show()
```

```python
df_workMode_prod=df.groupby('work_mode')['productivity_0_100'].mean().reset_index()
sns.barplot(data=df_workMode_prod,x='work_mode',y='productivity_0_100',hue='work_mode',palette='Blues')
plt.xlabel('Work Mode')
plt.ylabel('Productivity Score(0-100)')
plt.ylim(0,60)
plt.title('Work Mode VS Productivity',fontsize=14)
plt.show()
```
