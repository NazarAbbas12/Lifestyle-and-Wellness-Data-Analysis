```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv('ScreenTime vs MentalWellness.csv')
```

```python
df.info()
```

```python
numeric_df=df.select_dtypes(include=['int64','float64'])
numeric_df=numeric_df.drop(columns=['Unnamed: 15'],errors='ignore')
```

```python
corr_matrix=numeric_df.corr()
corr_matrix
```

```python
label=['Age','Screen Time Hours','Work Screen Hours','Leisure Screen Hours',
        'Sleep Hours','Sleep Quality(1-5)','Stress Level(0-10)','Productivity Level(0-100)',
        'Exercise per week(minutes)','Social Hours(week)','Mental Wellness Index(0-100)']
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',center=0,fmt='.2f',
            xticklabels=label,yticklabels=label)
plt.title("Correlation Heatmap of Lifestyle and Wellness Data", fontsize=14, pad=15)
plt.xticks(rotation=80)
plt.show()
```

 Correlation Insights from the Heatmap
 Key Observations

1.  Stress Level vs Mental Wellness  -0.91
    Very strong negative correlation: Higher stress levels drastically reduce mental wellness.

2.  Stress Level vs Productivity  -0.88
    Increased stress significantly lowers productivity  stress management is essential for better output.

3.  Sleep Quality vs Mental Wellness  +0.75
    Better sleep quality is strongly linked to higher mental wellness and emotional stability.

4.  Productivity vs Mental Wellness  +0.90
    Productive individuals tend to report better mental wellness, likely due to positive routines and balance.

5.  Screen Time vs Stress Level  +0.70
    More screen time is associated with higher stress  digital overload may negatively impact mental health.

6.  Screen Time vs Productivity  -0.71
    Excessive screen time lowers productivity, especially when dominated by non-work activities.

7. Sleep Hours vs Sleep Quality  +0.61
    Longer sleep duration generally improves quality, but consistency and restfulness matter more.

8.  Exercise per Week vs Mental Wellness  +0.23
    Exercise has a mild positive effect  even small amounts help boost wellness.

9.  Social Hours vs Mental Wellness  +0.07
    Social activity shows a weak positive correlation  slight improvement in wellness with interaction.

 Overall Insight

The heatmap reveals that stress, sleep quality, and screen time are the most influential factors for both productivity and mental wellness.
Maintaining balance in these areas can significantly enhance overall well-being and performance.
